```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.patches as patches

plt.style.use('seaborn-v0_8-dark-palette')
fig, ax = plt.subplots(figsize=(12, 7), facecolor='whitesmoke')

population_age = [22, 55, 62, 45, 21, 22, 34, 42, 42, 4, 2, 102, 95, 85, 55, 110, 120, 70, 65, 55, 111, 115, 80, 75, 65, 54, 44, 43, 42, 48]
bins = np.arange(0, 130, 10)

n, bins, patches_list = ax.hist(population_age, bins=bins, edgecolor='black', rwidth=0.85)

colors = plt.cm.viridis(np.linspace(0, 1, len(patches_list)))
for patch, color in zip(patches_list, colors):
    patch.set_facecolor(color)
    patch.set_linewidth(1.2)
    patch.set_linestyle('--')
    patch.set_alpha(0.8)
    patch.set_capstyle('round')

ax.set_title('Population Age Distribution', fontsize=22, fontweight='bold', color='darkblue')
ax.set_xlabel('Age Groups', fontsize=16)
ax.set_ylabel('Number of People', fontsize=16)
ax.set_xticks(bins)
ax.set_xticklabels([f'{int(b)}-{int(b+9)}' for b in bins], fontsize=13, rotation=45)
ax.set_yticks(np.arange(0, max(n)+2, 2))
ax.tick_params(axis='y', labelsize=13)
ax.grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)

t1 = np.arange(0.0, 5.0, 0.1)
t2 = np.arange(0.0, 5.0, 0.02)

plt.subplot(121)
plt.plot(t1, f(t1), 'bo', label='f(t) = exp(-t) * cos(2*pi*t)')
plt.plot(t2, f(t2), label='Smooth Line')
plt.title('Graph 1: Exponential Decay with Cosine')
plt.legend()

plt.subplot(122)
plt.plot(t2, np.cos(2*np.pi*t2), 'r', label='cos(2*pi*t)')
plt.title('Graph 2: Cosine Wave')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)

t1 = np.arange(0.0, 5.0, 0.1)
t2 = np.arange(0.0, 5.0, 0.02)

plt.subplot(121)
plt.plot(t1, f(t1), 'g^', label='f(t) = exp(-t) * cos(2*pi*t)', markersize=8)
plt.plot(t2, f(t2), 'b-', label='Smooth Line', linewidth=2)
plt.title('Graph 1: Exponential Decay with Cosine', fontsize=14, fontweight='bold')
plt.legend()

plt.subplot(122)
plt.plot(t2, np.cos(2*np.pi*t2), 'r--', label='cos(2*pi*t)', linewidth=2)
plt.title('Graph 2: Cosine Wave', fontsize=14, fontweight='bold')
plt.legend()

plt.tight_layout()
plt.show()
```
